### 1. 통계적 방법
가장 많이 사용되는 표준어를 추천해주는 방법
<img src='./image/통계적방법_1.png' width='70%'>

#### 1.1. 수행방법
1. 첫번째 단계 : 단어 사전 구축 하기
 - 표준 단어 사전을 받아서 빈도를 1로 초기화해서 데이터베이스에 업로드 한다.
2. 두번째 단계 : 입력 문자에 해당하는 표준 단어 리스트업
 - 사용자가 입력하는 문자에 해당하는 표준 단어 리스트를 빈도, 단어순으로 보여준다.
 - 사용자가 검색을 진행하면 검색된 단어의 빈도를 증가 시켜 추후 표준 단어 리스트에 상위에 노출되도록 한다.
 
#### 1.2. 문제점
- 표준단어에 없는 단어의 경우 추천이 어려움
- 추천은 해주지만 잘못 적은 단어에 대한 수정을 추천해주지 않음

### 2. 인공지능(Peter Norvig’s 알고리즘 : 피터 노빙 알고리즘)
초기 인공지능은 모든 경우의 수를 컴퓨터가 계산한다는 방향에서 시작됨.
사용자가 검색한 단어가 표준 단어의 경우 그대로 사용하지만 표준 단어가 아닌 경우 해당 단어로 만들 수 있는 모든 표준 단어를 만들어서 추천

#### 2.1. 수행방법
1. 첫번째 단계 : 단어 사전 구축 하기
 - 표준 단어 사전을 받아서 빈도를 1로 초기화해서 데이터베이스에 업로드 한다.
 - 단어가 빈도는 오타 교정 후보 단어가 여러개 나온 경우, 교정 단어의 우선순위를 정할때 사용한다.
2. 두번째 단계 : 키워드 교정하기
 - 검색 키워드를 오타교정 함수
 - 단어 사전에 있는 키워드인 경우, 그대로 키워드를 반환
 - 단어사전에서 키워드를 찾지 못한 경우, 검색 키워드의 편집거리가 1인 키워드 목록을 생성한다. 그리고 이 키워드가 단어사전에 있으면 반환
 - 편집거리 1인 키워드 목록에서도 표준 단어를 찾지 못하는 경우, 편집거리가 2인 키워드 목록을 구한 후, 3번째 단계와 동일하게 단어 사전을 찾는다. 
 - 만약 4번째 단계에서도 단어를 찾지 못한다면, 원본 키워드를 반환
 - 피터 노빙 알고리즘의 후보군 : deletes, transposes + replaces + inserts 과정을 거침

<img src='./image/피터노빙.png' width='60%'>
 
#### 2.2. 문제점
- 편집거리 2에 해당하는 후보군 단어의 종류가 너무 많이 나온다. (고성능 컴퓨팅 파워 or 오랜시간의 계산 과정이 걸릴 것으로 예상됨)
- 표준 단어 사전에 없는 단어로는 교정이 불가능하다.
---

### 3. 딥러닝
인공지능의 방법인 모든 가능한 후보군을 가지고 계산하는 방법은 많은 후보군에 의한 컴퓨팅 파워가 필요할 것으로 보여서 RNN 딥러닝을 통한 방법을 고려하고 있다. 아직까지는 정확히 어떤 알고리즘을 가지고 구성할지 생각해보지 못하였다.
