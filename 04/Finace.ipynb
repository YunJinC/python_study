```python
import pandas_datareader as pdr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class Finace:
    def getStocks(self, stock='FB', site='yahoo'):
        return pdr.DataReader(stock, site)
```

```python
f = Finace()
df = f.getStocks()
df.unstack(level=0)['Close'].plot(subplots=True)
```
